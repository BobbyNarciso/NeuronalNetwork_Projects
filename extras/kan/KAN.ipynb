# AproximaciÃ³n con Kolmogorov-Arnold Networks usando Redes Neuronales Densas

Este proyecto implementa un modelo de red neuronal **densa** para aproximar datos generados a partir de una funciÃ³n no lineal con ruido aleatorio. La implementaciÃ³n utiliza la **API funcional de Keras** para mayor claridad y flexibilidad.

## **Resumen del CÃ³digo**

1. **GeneraciÃ³n de Datos SintÃ©ticos**:
   - Se generan \( x \) aleatorios en el intervalo \([0, 1]\) con dimensiÃ³n **10**.
   - La salida \( y \) es calculada como:
     \[
     y = \sin\left(\sum x\right) + \text{ruido aleatorio}
     \]
     donde el ruido sigue una distribuciÃ³n normal de media cero y varianza pequeÃ±a.

2. **ConstrucciÃ³n del Modelo**:
   - Se implementa un modelo de red neuronal densa usando la **API funcional** de Keras.
   - Arquitectura del modelo:
     - **Entrada**: Vector de dimensiÃ³n \( 10 \).
     - **Capas Ocultas**:
       - 1Âª capa densa: 64 unidades, activaciÃ³n **ReLU**.
       - 2Âª capa densa: 32 unidades, activaciÃ³n **ReLU**.
       - 3Âª capa densa: 16 unidades, activaciÃ³n **ReLU**.
     - **Capa de salida**: 1 unidad, activaciÃ³n lineal.
   - La salida es un valor continuo aproximado.

3. **CompilaciÃ³n y Entrenamiento**:
   - OptimizaciÃ³n con el algoritmo **Adam**.
   - FunciÃ³n de pÃ©rdida: **Error CuadrÃ¡tico Medio (MSE)**.

     - NÃºmero de Ã©pocas: **200**.
     - TamaÃ±o del batch: **32**.

**Modelado de la red neuronal KAN** 

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sintÃ©ticos
def generate_data(num_samples=1000, input_dim=10):
    x = np.random.rand(num_samples, input_dim).astype(np.float32)
    y = np.sin(np.sum(x, axis=1)) + 0.1 * np.random.randn(num_samples)
    return x, y.reshape(-1, 1)

# Construir el modelo usando Input y capas Dense
def build_model(input_dim):
    inputs = tf.keras.Input(shape=(input_dim,))
    x = tf.keras.layers.Dense(64, activation="relu")(inputs)
    x = tf.keras.layers.Dense(32, activation="relu")(x)
    x = tf.keras.layers.Dense(16, activation="relu")(x)
    outputs = tf.keras.layers.Dense(1)(x)
    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model

# Entrenamiento y evaluaciÃ³n
def train_model():
    input_dim = 10
    x_train, y_train = generate_data(num_samples=1000, input_dim=input_dim)

    model = build_model(input_dim)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss="mse")

    # Entrenamiento del modelo
    model.fit(x_train, y_train, epochs=200, batch_size=32, verbose=1)

    # PredicciÃ³n
    y_pred = model.predict(x_train)

    # Graficar resultados
    plt.figure(figsize=(10, 6))
    plt.scatter(range(len(y_train)), y_train[:, 0], label="Datos reales", alpha=0.7)
    plt.scatter(range(len(y_pred)), y_pred[:, 0], label="Predicciones", alpha=0.7)
    plt.title("AproximaciÃ³n con Red Neuronal densa")
    plt.xlabel("Ãndice de datos")
    plt.ylabel("Valor")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    train_model()
```
